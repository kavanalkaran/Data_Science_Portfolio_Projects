```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r"C:\Users\kavan\OneDrive\Desktop\EVIndia.csv")
```

```python
df1
```

```python
# Exploratory Data Analysis for Electric Vehicle Price Prediction Data

#filling empty data with 0

df1 = df1.fillna('0')
```

```python
df1
```

```python
# Now, let us get the name of the car manufacturer from the car name.

df1['Manufacturer'] = df1.Car.str.split(' ', 1, expand=True)[0]
```

```python
df1
```

```python
print(df1['Manufacturer'])
```

```python
#cleaning the data 
#removing  the currency sign

PriceRange = []
for item in df1["PriceRange"]:
    PriceRange+=[int(item.average(PriceRange))]
df1["PriceRange"] = PriceRange

# ask about dash and how to make it one int
```

```python
Range = []
for item in df1['Range']:
    Range+=[int(item.replace('Km/Full Charge','').replace('na','0'))]
df1['Range'] = Range
```

```python
BootSpace = []
for item in df1["BootSpace"]:
    BootSpace+=[int(item.replace('L','').replace('na','0'))]
df1['BootSpace'] = BootSpace
```

```python
Capacity = []
for item in df1['Capacity']:
    Capacity+=[int(item.replace('Seater',''))]
df1['Capacity'] = Capacity
```

```python
BaseModel = []
for item in df1['BaseModel']:
    BaseModel += [str(item.replace('na','0'))]
df1['BaseModel'] = BaseModel
```

```python
TopModel = []
for item in df1['TopModel']:
    TopModel += [str(item.replace('na','0'))]
df1['TopModel'] = TopModel
```
